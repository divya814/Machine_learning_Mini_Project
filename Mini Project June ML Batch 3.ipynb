 ## Q4. Which are the movies with most and least earned revenue?

```python
def find_minmax(x):    
    #use the function 'idxmin' to find the index of lowest profit movie.    
    min_index = data[x].idxmin()
    
    #use the function 'idxmax' to find the index of Highest profit movie.
    high_index = data[x].idxmax()
    
    high = pd.DataFrame(data.loc[high_index,:])
    low = pd.DataFrame(data.loc[min_index,:])
    print("Movie with Highest "+ x + " : ",data['original_title'][high_index])
    print("Movie with Lowest "+ x + "  : ",data['original_title'][min_index])
    
    return pd.concat([high,low],axis = 1)

find_minmax('revenue')
```

## Ans. The movie with Highest revenue is Avatar and  the movie with lowest revenue  is Wild Card

##  Q6. Name any 3 production companies which have invested money in worse revenue movies?
